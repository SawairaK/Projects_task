**Sum of all Even Numbers**

```python
# Initialize a variable to store the sum of even numbers
sum = 0

# Loop through numbers from 1 to 100
for i in range(100):
    # Check if the number is even
    if i % 2 == 0:
        # Add the even number to the sum
        sum += i

# Print the sum of all even numbers
print("The sum of all even numbers from 1 to 100 is:", sum)
```

```python
# Input from the user
num = int(input("Enter a number: "))

# Prime numbers are greater then 1
if num > 1:
    for i in range(2, num):
        if (num % i) == 0:
            print(num, "is not a prime number")
            break
    else:
        print(num, "is a prime number")
else:
    print(num, "is not a prime number")
```

**program that reverses a given string without using any built-in reverse function.**

```python
def reverse_string(string):
    rev_string = ""
    for i in string:
        rev_string = i + rev_string
    return rev_string

string = "HelloWorld"
print("The original string is : ",string)
reverse = reverse_string(string)
print("The reversed string  is : ",reverse)

```

**a function that finds the factorial of a given number using a loop.**

```python
def factorial(number):
    # Initialize the result to 1 (factorial of 0 and 1 is 1)
    factorial = 1

    # Loop through each number from 1 to n
    for i in range(1, n + 1):
        factorial *= i  # Multiply result by the current number

    return factorial

# Example usage
number = input("Enter a number:")
factorial_of_number = factorial(number)
print(f"The factorial of number :" ,factorial_of_number)
```

**counts the number of vowels in a string**

```python
def count_vowels(string):
    # Define a set of vowels
    vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']

    # Initialize the count
    count = 0

    for char in string:
        if char in vowels:
            count += 1

    return count
# Input a string from the user
string = input("Enter a string: ")
count = count_vowels(string)
print(f"The number of vowels in the string is:",count)
```

**function that checks if a given word is a palindrome.**

```python
def palindrome(word):
    # Convert the word to lowercase to make the check case-insensitive
    word = word.lower()

    # Reverse the word
    #[::-1] slice notation used to reverse the string. It consists of three parts: start:stop:step
    reversed_word = word[::-1]

    # Check if the word is the same as the reversed word
    return word == reversed_word

# Input a word from the user
word = input("Enter a word: ")
word_is_palindrome = palindrome(word)

if word_is_palindrome:
    print("The word is a palindrome.")
else:
    print("The word is not a palindrome.")
```

**Fibonacci numbers up to a specified limit.**

```python
def fibonacci_series(number):
    # Initialize the Fibonacci sequence with the first two numbers
    fib_sequence = [0, 1]  #list
    count=1
    # Continue generating Fibonacci numbers until the next number exceeds the limit
    while (count<=number):
        next_fib = fib_sequence[-1] + fib_sequence[-2]
        if next_fib > number:
            break
        fib_sequence.append(next_fib)

    return fib_sequence

number = int(input("Enter the number for Fibonacci numbers: "))

# Generate the Fibonacci sequence up to the limit
fibonacci_list = fibonacci_series(number)

# Print the generated Fibonacci sequence
print(f"Fibonacci numbers up to {number}: {fibonacci_list}")
```

**finds the intersection of two lists.**

```python
def find_intersection(list1, list2):
    # Convert both lists to sets and find their intersection using and operator (both true)
    intersection = list(set(list1) & set(list2))

    return intersection

list1 = [1, 2, 3, 4, 5,6,9]
list2 = [4, 5, 6, 7, 8,6,0]

# Find the intersection of the two lists
intersection_list = find_intersection(list1, list2)

# Print the result
print("The intersection of the lists is:",intersection_list)
```

**replaces all occurrences of a specific word in a text with another word.**

```python
def replace_word(text, old_word, new_word):
    # Replace all occurrences of old_word with new_word
    replaced_text = text.replace(old_word, new_word)
    return replaced_text

# Input text from the user
text = input("Enter text: ")
old_word = input("Enter the word to be replaced: ")
new_word = input("Enter the replacement word: ")

# Replace the word and print the result
replaced_text_word = replace_word(text, old_word, new_word)
print("Updated text:", replaced_text_word)
```

**maximum and minimum values in a tuple of numbers.**

```python
def maximum_minimum(numbers):
   # Use the built-in functions to find the maximum and minimum values
    max_value = max(numbers)
    min_value = min(numbers)

    return max_value, min_value
    # Use loop to find the maximum and minimum values
    for num in numbers:
        if num > max_value:
            max_value = num
        if num < min_value:
            min_value = num
    return max_value, min_value
tuple_number = (22, 1, 45, 11, 70, 34, 55)

# Find the maximum and minimum values
max_value, min_value = maximum_minimum(tuple_number)

# Print the results
print(f"The maximum value is: {max_value}")
print(f"The minimum value is: {min_value}")
```

**sorts a list of strings in alphabetical order.**

```python
def sort(strings_list):
    n = len(strings_list)

    # Iterate over each element in the list
    for i in range(n):
        # Compare each element with the remaining elements
        for j in range(i + 1, n):
            # If the current element is greater than the next one, swap them
            if strings_list[i] > strings_list[j]:
                strings_list[i], strings_list[j] = strings_list[j], strings_list[i]

str_list = ['hello', 'world', 'control', 'descendent', 'java', 'python']
sort(str_list)

print("Sorted List:", str_list)
```

**list of integers and returns a new list with unique values**

```python
def remove_duplicates(numbers):
    unique_numbers = []
    for number in numbers:
        if number not in unique_numbers:
            unique_numbers.append(number)
    return unique_numbers


input_string = input("Enter a list of integers separated by spaces: ")
# Split the String into a List:
string_list = input_string.split()
# Convert Strings to Integers:
numbers_list = []
for item in string_list:
    numbers_list.append(int(item))

unique_list = remove_duplicates(numbers_list)

print("List with unique values:", unique_list)
```
