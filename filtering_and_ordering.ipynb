```python
import pandas as pd
```

```python
df = pd.read_csv('/content/world_population.csv')
df
```

```python
# Filter based on columns
df[df['Rank'] <=10 ]
```

```python
# If we look for specific information use isin() Function
specific_countries = ['Bangladesh','Brazil']
df[df['Country'].isin(specific_countries)]
```

```python
# If we search for specific str value or number
df[df['Country'].str.contains('United')]
```

```python
# Filtering based on indexs
df = df.set_index('Country')
df
```

```python
# loc (location) or iloc (integer location) axis = 0 (rows) asix = 1 (column)
df.filter(items = ['Continent','CCA3'], axis = 1)
```

```python
df.filter(items = ['Zimbabwe'], axis = 0)
```

```python
# like keyword give all the information with the word like in data
df.filter(like = 'United', axis = 0)
```

```python
# loc function give the specific value
df.loc['United States']
```

```python
# lets do order by (sorting)
df
```

```python
# We want to sort Rank column in Ascending order
df[df['Rank'] < 10].sort_values(by = 'Rank',ascending=True)
```

```python
# If we want to sort multiple columns we create list
df[df['Rank'] < 10].sort_values(by=['Continent','Country'],ascending=[False,True])
```
